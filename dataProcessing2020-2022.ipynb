```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python

folder_path = 'data/2020'  
file_extension = '*.xlsx'  

# Get a list of all Excel files in the folder
excel_files = glob.glob(f'{folder_path}/{file_extension}')



```

```python
desired_columns = [
    'Water Company Name', 
    'Site Name',
    'Shellfishery      (Only populate for storm overflows that have Shellfish water EDM requirements)',
    'Bathing Beach (Only populate for storm overflows that have bathing water EDM requirements)',
    'Total Duration (hours) of all spills prior to processing through 12-24 hour counting method',
    'Counted spills using 12-24hr counting method',
    '% of reporting period EDM operational'
]
```

```python
# Create an empty DataFrame to store the combined data
combined_df = pd.DataFrame()

# Iterate over each Excel file
for excel_file in excel_files:
    
    excel_data = pd.ExcelFile(excel_file)
    sheet_names = excel_data.sheet_names

    # Iterate over each sheet in the current Excel file
    for sheet_name in sheet_names:
        sheet_df = pd.read_excel(excel_file, sheet_name=sheet_name)
        filtered_df = sheet_df[desired_columns]
        # Append the DataFrame to the combined DataFrame
        combined_df = combined_df._append(filtered_df, ignore_index=True)
        
    
#combined_df = combined_df.drop('Unnamed: 10', axis=1)
print(combined_df)

df_2020 = combined_df
```
